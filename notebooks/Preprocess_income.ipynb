```python
# import packages
import pandas as pd
import numpy as np
import os
```

```python
# set os
main_directory = os.getcwd().strip('notebooks')
```

```python
# read in data
link = 'https://www.abs.gov.au/statistics/labour/earnings-and-working-conditions/personal-income-australia/2014-15-2018-19/6524055002_DO001.xlsx'
pdf = pd.read_excel(link, 'Table 1.4')
pdf

```

```python
#clean df remove redundent headers and replace with new headers
header = pdf.iloc[5]
new_df = pdf.iloc[6:]
new_df.columns = header
new_df
```

```python
#convert sa2 codes to string for use of regex
new_df['SA2'] = new_df['SA2'].astype('string')
```

```python
#use regex to filter SA2 locations in victoria
#SA2 Maincode made up of 9 digits & first digit represents the state, where VIC=2
regex='2[0-9]{8}'
vic_df=new_df[new_df.SA2.str.match(regex)]
#reset index and remove last column 
vic_df = vic_df.reset_index(drop=True).drop(vic_df.columns[-1], axis=1)
vic_df
```

```python
# Rename headers
vic_df.set_axis([ 'SA2', 'SA2 NAME', 'Earners 2015', 'Earners 2016', 'Earners 2017', 'Earners 2018', 'Earners 2019', 'Median Age 2015', 'Median Age 2016', 'Median Age 2017', 'Median Age 2018', 'Median Age 2019', \
'Total $ 2015', 'Total $ 2016', 'Total $ 2017', 'Total $ 2018', 'Total $ 2019', 'Median $ 2015', 'Median $ 2016', 'Median $ 2017', 'Median $ 2018', 'Median $ 2019', 'Mean $ 2015', 'Mean $ 2016', 'Mean $ 2017', 'Mean $ 2018', 'Mean $ 2019'], axis='columns', inplace=True)
```

```python
vic_df.to_csv(main_directory + 'data/curated/vic_income_by_sa2.csv')
```
