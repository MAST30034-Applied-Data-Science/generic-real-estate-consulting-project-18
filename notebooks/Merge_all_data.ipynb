```python
# import packages
import os
import pandas as pd
```

```python
# set directory
main_directory=os.getcwd().strip('notebooks')
```

## Generate modelling dataset

```python
# get rental_population df
rental_pop=pd.read_csv(main_directory+'data/curated/population/rental_population.csv')
```

```python
rental_pop
```

```python
# get conversion data
conversion = pd.read_csv(main_directory+'data/curated/conversion/LGA_to_SA2.csv')
```

```python
# get crime data
crime =pd.read_csv(main_directory+'data/curated/crime/crime_ratios_all.csv')
```

```python
# drop unneeded columns
for i in range(12,21):
    rental_pop.drop(columns = ['pop_20' + str(i)], inplace = True)
```

```python
rental_pop.drop(columns = ['Unnamed: 0', 'pop_density_2021', 'S/T name','SA2 code_x'], inplace = True)
```

```python
# get income data - current
current_income = pd.read_csv(main_directory + 'data/curated/income/vic_income_by_sa2.csv')
# only want most recent income - mean and median, whichever one works
current_income_relevant = pd.DataFrame()
current_income_relevant['SA2 NAME']=current_income['SA2 NAME']
current_income_relevant['Median $ 2019'] = current_income['Median $ 2019']
current_income_relevant['Mean $ 2019'] = current_income['Mean $ 2019']
```

```python
# get current crime data
current_crime = pd.DataFrame()
current_crime['LGA code'] = crime['LGA code']
current_crime['2021']= crime['2021']
```

```python
# get population data
population = pd.read_csv(main_directory+'data/curated/population/by age/SA2_Vic_age_clean.csv')
```

```python
# rename columns so its identifiable
population = population.add_suffix('_population')
```

```python
# merge population and rental
merge_pop_rent = rental_pop.merge(population,how = 'left', left_on = ['SA2 name'], right_on=['SA2 name_population'])
```

```python
# drop unnecessary columns
merge_pop_rent.drop(columns = ['Unnamed: 0_population', 'SA2 code_y'], inplace = True)
```

```python
merge_pop_rent
```

```python
# put indexing columns at the front
SA2code = merge_pop_rent['SA2 code_population']
merge_pop_rent.drop(columns = ['SA2 code_population'], inplace = True)
merge_pop_rent.insert(2,'SA2 code',SA2code)
```

```python
# get the LGAs into the dataset for inputing crime
merge_LGA = merge_pop_rent.merge(conversion, how = 'left', left_on = ['SA2 name'], right_on = ['SA2_NAME_2021'])
```

```python
# merge crime and the rest of the data
merge_crime_all = merge_LGA.merge(current_crime, how = 'left', left_on = ['LGA_NAME_2021'], right_on = ['LGA code'])
```

```python
# drop unnecessary columns
merge_crime_all.drop(columns = ['SA2 name_population','Unnamed: 0', 'SA2_CODE_2021', 'SA2_NAME_2021', 'LGA_CODE_2021', 'LGA_NAME_2021'], inplace=True)
```

```python
merge_crime_all.rename(columns = {"2021": "2021 crime rate"},inplace = True) # rename crime rate col
```

```python
# put indexing columns at the front
LGAcode = merge_crime_all['LGA code']
merge_crime_all.drop(columns = ['LGA code'], inplace = True)
merge_crime_all.insert(2,'LGA code',LGAcode)
```

```python
# merge with current income -> completed all non poi merging
merge_all = merge_crime_all.merge(current_income_relevant, how = 'left',left_on = ['SA2 name'], right_on = ['SA2 NAME'])
```

```python
merge_all.drop(columns = ["SA2 NAME"], inplace = True)
```
