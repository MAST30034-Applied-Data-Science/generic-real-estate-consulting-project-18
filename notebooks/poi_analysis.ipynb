```python
category_types = [] # Unique category types
count_dict = defaultdict(dict) # Store counts of category types for each instance
null_dict = defaultdict(dict) # Store filename and indexes of null values

# Iterate through each POI .json file
for elem in filelist:
    with open(f'{PATH}{elem}', 'r') as f:
        file = load(f)

    # Iterate through each instance and check for NaN values.
    # If not Nan, get a count of POI category types for each property
    # as well as a running list of unique POI category types seen
    for i in range(len(file)):
        property_coordinates = list(file[i].keys())[0] # Coordinates of property
        iter_features = list(file[i].values())[0] # List of POIs for property
        num_features = len(iter_features) # Number of POIs for that property

        
        if type(iter_features[0]) == list:
            # Not null, iterate through features of
            # a property and perform preprocessing
            
            count_dict[property_coordinates]['total'] = 0
            for j in range(num_features):
                
                category = iter_features[j][-1]
                category_name = category['category_name']
                
                # If category_name is relevant, append count data
                if category_name in categories:
                    if category_name not in category_types:
                        category_types.append(category['category_name'])
                    
                    # Get counts
                    count_dict[property_coordinates][category_name] = count_dict \
                        .get(property_coordinates, dict()) \
                            .get(category_name, 0) +1
                            
                    count_dict[property_coordinates]['total'] += 1

        else:
            # Null, save index for future reference
            print(f'Null at index {i} in document {elem}')
            null_dict[elem][i] = property_coordinates

    
    

            
```

```python
# Dump counts data to .json file
with open(f'../data/curated/poi_counts.json', 'w') as f:
        dump(count_dict, f, indent=4, separators=(',',': '))
```

```python
# Save category types to text file
with open("../data/curated/poi_types.txt", "w") as fp:
    for item in category_types:
        fp.write("%s\n" % item)
```
