```python
import pandas as pd
import numpy as np
```

```python
pop_SA2 = pd.read_csv("generic-real-estate-consulting-project-18/data/raw/Population_estimate_2001-2021_partial_cleaned.csv")
pop_LGA = pd.read_csv("generic-real-estate-consulting-project-18/data/raw/LGA_population 2001-2021 partially cleaned.csv")
LGA_lookup = pd.read_csv("generic-real-estate-consulting-project-18/data/raw/LGA_2020_VIC.csv")
```

```python
pop_SA2_vic = pop_SA2[pop_SA2['S/T name'] == 'Victoria']
pop_SA2_vic
```

```python
pop_SA2_vic.drop(pop_SA2_vic.iloc[:, 2:8], inplace=True, axis=1)
```

```python
pop_SA2_vic.drop(pop_SA2_vic.iloc[:, 4:15], inplace=True, axis=1)
```

```python
pop_SA2_vic.drop(['No.change2011-2021', '%change2011-2021'], axis = 1)
```

```python
pop_LGA.drop(pop_LGA.iloc[:, 2:13], inplace=True, axis=1)
```

```python
pop_LGA.drop(pop_LGA.iloc[:, 12:16], inplace=True, axis=1)
```

```python
pop_LGA = pop_LGA.dropna()
```

```python
pop_LGA['LGA code'] = pop_LGA['LGA code'].astype('int')
```

```python
LGA_lookup = LGA_lookup.drop(['MB_CODE_2016', 'AREA_ALBERS_SQKM'], axis = 1)
```

```python
LGA_lookup = LGA_lookup.drop_duplicates()
```

```python
LGA_lookup = LGA_lookup.drop(LGA_lookup.index[0])
LGA_lookup = LGA_lookup.drop(LGA_lookup.index[-1])
```

```python
LGA_lookup.rename(columns = {'LGA_CODE_2020':'LGA code'}, inplace = True)
```

```python
LGA_full = pd.merge(pop_LGA, LGA_lookup, how="inner", on=["LGA code"])
```

```python
LGA_full.drop(LGA_full.iloc[:, 14:18], inplace=True, axis=1)
```
