```python
import pandas as pd
import numpy as np
import os
```

```python
# set main directory
main_directory = os.getcwd().strip('notebooks')
```

```python
# loads the population data for both SA2 and LGA

pop_SA2 = pd.read_csv(main_directory+"data/raw/SA2_Population 2001-2021 partial_cleaned.csv")
pop_LGA = pd.read_csv(main_directory+"data/raw/LGA_population 2001-2021 partially cleaned.csv")
LGA_lookup = pd.read_csv(main_directory+"data/raw/LGA_2020_VIC.csv")
```

```python
# filters out the dataframe to only have SA2 in Victoria

pop_SA2_vic = pop_SA2[pop_SA2['S/T name'] == 'Victoria']
pop_SA2_vic
```

```python
# drops the irrelevant columns which are GCCSA,SA3 and SA4

pop_SA2_vic.drop(pop_SA2_vic.iloc[:, 2:8], inplace=True, axis=1)
```

```python
# drops years pre 2012 as we want a 10 year period

pop_SA2_vic.drop(pop_SA2_vic.iloc[:, 4:15], inplace=True, axis=1)
```

```python
# drops the percentage change and number of change from 2011-2021

pop_SA2_vic = pop_SA2_vic.drop(['No.change2011-2021', '%change2011-2021'], axis = 1)
```

```python
pop_SA2_vic.rename(columns = {'2012':'pop_2012',
                             '2013':'pop_2013',
                             '2014':'pop_2014',
                             '2015':'pop_2015',
                             '2016':'pop_2016',
                             '2017':'pop_2017',
                             '2018':'pop_2018',
                             '2019':'pop_2019',
                             '2020':'pop_2020',
                             '2021':'pop_2021',
                             'Area/km2': 'Area',
                             'Population density 2021/km2':'pop_density_2021'
                             }, inplace = True)
```

```python
pop_LGA
```

```python
# drops years 2001-2011

pop_LGA.drop(pop_LGA.iloc[:, 2:13], inplace=True, axis=1)
```

```python
# drops the unnamed columns and the 2011-2021 change columns

pop_LGA.drop(pop_LGA.iloc[:, 12:16], inplace=True, axis=1)
```

```python
# drops NaN values in the dataframe

pop_LGA = pop_LGA.dropna()
```

```python
# converts the LGA code column to int for merge

pop_LGA['LGA code'] = pop_LGA['LGA code'].astype('int')
```

```python
# drops the irrelevant columns for the LGA lookup dataframe

LGA_lookup = LGA_lookup.drop(['MB_CODE_2016', 'AREA_ALBERS_SQKM'], axis = 1)
```

```python
# drops duplicates as the dataframe is filled with tons of them

LGA_lookup = LGA_lookup.drop_duplicates()
```

```python
# drops first and last rows as they dont exist in LGA

LGA_lookup = LGA_lookup.drop(LGA_lookup.index[0])
LGA_lookup = LGA_lookup.drop(LGA_lookup.index[-1])
```

```python
# renamed column for merging purposes

LGA_lookup.rename(columns = {'LGA_CODE_2020':'LGA code'}, inplace = True)
```

```python
# merge the LGA lookup with the LGA population column

LGA_full = pd.merge(pop_LGA, LGA_lookup, how="inner", on=["LGA code"])
```

```python
# drops columns from the LGA lookup as they exist already in the LGA pop column

LGA_full.drop(LGA_full.iloc[:, 14:18], inplace=True, axis=1)
```

```python
# exports these dataframes into a csv file

LGA_full.to_csv(main_directory+ "data/curated/LGA_Vic_pop_clean.csv")
pop_SA2_vic.to_csv(main_directory+"data/curated/SA2_Vic_pop_clean.csv")
```
