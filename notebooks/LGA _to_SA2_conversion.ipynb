```python
import pandas as pd
import glob
import os
import numpy as np
```

```python
# set main
main_directory = os.getcwd().strip('notebooks')
```

```python
SA2_lookup = pd.read_csv(main_directory+"data/raw/MB_2021_AUST_lookup.csv")
LGA_lookup_2021 = pd.read_csv(main_directory+"data/raw/LGA_2021_AUST_lookup.csv")
```

```python
SA2_lookup = SA2_lookup[SA2_lookup['STATE_NAME_2021'] == 'Victoria']
SA2_lookup
```

```python
SA2_lookup = SA2_lookup[SA2_lookup.AREA_ALBERS_SQKM.notnull()]
```

```python
SA2_lookup
```

```python
dropped_columns = ['MB_CATEGORY_2021', 'SA1_CODE_2021',
                   'SA3_CODE_2021','SA3_NAME_2021', 'SA4_CODE_2021', 'SA4_NAME_2021', 
                   'GCCSA_CODE_2021', 'GCCSA_NAME_2021', 'AUS_CODE_2021', 'AUS_NAME_2021', 
                   'AREA_ALBERS_SQKM', 'ASGS_LOCI_URI_2021', 'CHANGE_FLAG_2021', 'CHANGE_LABEL_2021']
SA2_lookup = SA2_lookup.drop(dropped_columns, axis=1)
```

```python
SA2_lookup
```

```python
print(len(SA2_lookup['SA2_CODE_2021'].unique()))
```

```python
SA2_lookup
```

```python
LGA_lookup_2021 = LGA_lookup_2021[LGA_lookup_2021.STATE_NAME_2021 == 'Victoria']
LGA_lookup_2021
```

```python
LGA_LOOKUP_2021 = LGA_lookup_2021[LGA_lookup_2021.AREA_ALBERS_SQKM.notnull()]
```

```python
LGA_LOOKUP_2021 = LGA_LOOKUP_2021[LGA_LOOKUP_2021.LGA_NAME_2021 != 'Unincorporated Vic']
```

```python
LGA_LOOKUP_2021.drop(LGA_LOOKUP_2021.iloc[:, 3:9], inplace=True, axis=1)
```

```python
LGA_LOOKUP_2021
```

```python
# the SA2 french island is removed
combine_df = pd.merge(SA2_lookup,LGA_LOOKUP_2021,how = 'inner',on='MB_CODE_2021')
```

```python
combine_df
```

```python
combine_df = combine_df.drop_duplicates(subset = ['SA2_CODE_2021'],keep = 'first')
```

```python
combine_df.sort_values('SA2_CODE_2021', ascending = True)
```

```python
combine_df = combine_df.drop(['MB_CODE_2021', 'STATE_CODE_2021', 'STATE_NAME_2021'], axis = 1)
```

```python
combine_df.rename(columns = {'SA2_CODE_2021' : 'SA2 code',
                            'SA2_NAME_2021' : 'SA2 name',
                            'LGA_CODE_2021' : 'LGA code',
                            'LGA_NAME_2021' : 'LGA name'})
```

```python
combine_df.to_csv(("data/curated/conversion/LGA_to_SA2.csv"))
```
